```python
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:.5f}'.format
```

```python
delivery_df = pd.read_csv("./delivery_raw.csv", sep = '\t')

delivery_df
```

```python
delivery_df.dropna(inplace=True)
delivery_df = delivery_df.reset_index(drop=True)
```

```python
delivery_df['created_at'] = pd.to_datetime(delivery_df['created_at'])
delivery_df['actual_delivery_time'] = pd.to_datetime(delivery_df['actual_delivery_time'])
delivery_df['delivery_duration'] = delivery_df['actual_delivery_time'] - delivery_df['created_at']
delivery_df['delivery_duration'] = delivery_df['delivery_duration'].dt.total_seconds()

delivery_df[['delivery_duration','estimated_store_to_consumer_driving_duration']]
```

```python
from sklearn.preprocessing import OneHotEncoder

print(delivery_df['store_primary_category'])
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_df = pd.DataFrame(encoder.fit_transform(delivery_df[['store_primary_category']]))

print(encoded_df.shape)
print(delivery_df.shape)
delivery_df = pd.concat([delivery_df, encoded_df], axis = 1)
```

```python
from sklearn.preprocessing import MinMaxScaler

removed_list = ['delivery_duration',
                'created_at',
                'actual_delivery_time',
                'store_primary_category',
               'estimated_store_to_consumer_driving_duration',
               'estimated_order_place_duration',
]

tot_data = delivery_df[delivery_df.columns.difference(removed_list)].to_numpy()
tot_label = delivery_df['delivery_duration'].to_numpy()
print(tot_data)
scaler = MinMaxScaler() 
scaler.fit(tot_data) 
tot_data = scaler.transform(tot_data)     
```

```python
from sklearn.model_selection import train_test_split


train_data, test_data, train_label, test_label = train_test_split(tot_data, tot_label, test_size=0.1, random_state=1000)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer
from sklearn.linear_model import Ridge
def weighted_rmse(y_true, y_pred):
     score = np.where(y_true > y_pred, 2 * (y_true - y_pred)**2, (y_true - y_pred)**2) / np.sum(y_true < y_pred)*2
     return np.sqrt(np.mean(score))

score = make_scorer(weighted_rmse, greater_is_better=False)

grid = GridSearchCV(estimator=Ridge(), param_grid={'alpha' : [0., 1.0, 10.0]},
                      scoring=score, cv=5)
grid.fit(train_data, train_label)
```

```python
predicted_label = grid.predict(test_data)

def rmse(y_true, y_pred) :
  return np.sqrt(np.mean((y_true-y_pred) ** 2))

def cal_under_predict(y_true, y_pred) :
  return np.sum(y_true > y_pred) / len(y_true)

print(rmse(test_label, predicted_label))
print(cal_under_predict(test_label, predicted_label))
```
